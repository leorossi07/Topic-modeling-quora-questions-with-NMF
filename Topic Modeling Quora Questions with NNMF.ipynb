#### Applying NNMF

> First, we create a Vectorized Document Term Matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
```

```python
dtm = tfidf.fit_transform(df['Question'])
```

> Let's separate the Questions into 30 topics

```python
from sklearn.decomposition import NMF
```

```python
nmf = NMF(n_components=30, random_state=42)
```

```python
# Just for better display
import warnings

warnings.filterwarnings("ignore")
```

```python
nmf.fit(dtm)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

for index,topic in enumerate(nmf.components_):
    print(f'THE TOP 15 WORDS FOR TOPIC #{index}')
    top_words = [tfidf.get_feature_names()[i] for i in topic.argsort()[-15:]]
    print(top_words)
    print('\n')
    
    wordcloud = WordCloud(background_color="white").generate(" ".join(top_words))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f'Topic #{index}')
    plt.show()
    print('\n\n')
```

```python
topic_results = nmf.transform(dtm)
```

```python
df['Topic_Index'] = topic_results.argmax(axis=1)
```

```python
topic_labels = {
    0: "Books and Movies",
    1: "College and Career",
    2: "Quora Questions",
    3: "Making Money Online/Tech",
    4: "Life and Meaning",
    5: "India/Politics",
    6: "Opinions/Beliefs",
    7: "US Politics (Trump/Clinton)",
    8: "Language Learning (English)",
    9: "Relationships/Dating",
    10: "General Recommendations",
    11: "India Demonetization",
    12: "Relationships/Mindset",
    13: "Communication/Language Skills",
    14: "Weight Loss/Diet",
    15: "Travel/Time",
    16: "Love and Relationships",
    17: "Suicide/Self-harm",
    18: "World Conflicts/Wars",
    19: "Computer Science/Tech Careers",
    20: "Making Money Online",
    21: "Career/Business",
    22: "Social Media/Account Security",
    23: "New Year's Resolutions/Job",
    24: "2016 US Election/God",
    25: "2016 US Election/World Concerns",
    26: "Sex/Relationships",
    27: "Engineering Careers/Interviews",
    28: "Work-Life Balance",
    29: "Learning Programming"
}
# Create a new column with topic labels
df['Topic_Label'] = df['Topic_Index'].map(topic_labels)

# Display the distribution of topics
print(df['Topic_Label'].value_counts())
```
